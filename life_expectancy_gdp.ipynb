```python
#Divide gdp by 1000
#Histogram of life expectancy
#Histogram of gdp
#Line graph of how both average change over time
#Scatter plot of how gdp increases life expectancy
#bar graph of top 5 and bottom 5

from matplotlib import pyplot as plt

import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math

life_expectancy = pd.read_csv('data/all_data.csv')
#Normalizing GDP to have reasonable digits by rounding 2 places and dividing by a thousand
life_expectancy["GDP"] = life_expectancy["GDP"].apply(
    lambda x: round(x/1000, 2))

#Condensing the average for all years for each country:

life_expectancy["GDP"] = life_expectancy["GDP"].apply(
    lambda x: round(x/1000, 2))


```

This is an exploration comparing countries data for GDP and life expectancy throughout the years. 

Main goals are:

1) See how GDP and life expectancy changed throughout the years

2) See if the increase in GDP correlates to the increase of life expectancy

```python
#Frequency distribution of life expectancies

plt.hist(life_expectancy['Life expectancy at birth (years)'],  bins=20)
plt.xlabel('Average age of death')
plt.ylabel('Number of countires')
plt.title('Frequency of Age of Death')
plt.show()
plt.clf()

print("Majority of countires have a high life expectancy, but there are a concerning amount on the lower end of age")
```

```python
#Frequency distribution of gdp

plt.hist(life_expectancy["GDP"],  bins=20)

plt.xlabel('GDP')
plt.ylabel('Number of countires')
plt.title('Frequency distribution of GDP/1000')
plt.show()
plt.clf()

print("Majority of countries have a low GDP, and the outliers are the ones with high gdp. Right skewed graph")
```

```python
plt.scatter(life_expectancy["GDP"], life_expectancy['Life expectancy at birth (years)'], color='pink', marker = 'o')
plt.xlabel('GDP')
plt.ylabel('Average Age of Death')
plt.title('Age of Death vs GDP')
```

After a bare minimum GDP, the age of death is in a reasonable age (70+). The GDP does not have to be 
extrordinarily high for a country to have high life expectancy

```python
plt.scatter(life_expectancy["Year"], life_expectancy['Life expectancy at birth (years)'], color = "blue")
plt.xlabel('Year')
plt.ylabel('Average age of Death')
plt.title('Life Expectancy Througout the years')
plt.show()
plt.clf()
```

There has been gradual improvement of life expectancy after 2008 for the early death countries, the long
expectancy countries remain the same

```python
plt.scatter(life_expectancy["Year"], life_expectancy['GDP'], color='purple')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP chanage over years')
```

For the GDP graph by year, GDP improved for the top earning and middle earning countries, low earning
countires remained the same. \n\n
Both life expectancy and GDP increased throughout the years, but
1) The life expectancy increased most for countries with the lowest to begin with
2) GDP increased most for countries with the highest GDP

```python
life_expectancy_gdp = life_expectancy.groupby('Country')["GDP"].mean().reset_index()
print(life_expectancy_gdp)

plt.bar(range(len(life_expectancy_gdp["Country"])), life_expectancy_gdp["GDP"])
plt.title("Average GDP by Country")
#create your ax object here
ax = plt.subplot()
ax.set_xticks(range(len(life_expectancy_gdp["Country"])))
ax.set_xticklabels(life_expectancy_gdp["Country"], rotation=30)
plt.show()
plt.clf()
```

```python
life_expectancy_le= life_expectancy.groupby('Country')["Life expectancy at birth (years)"].mean().reset_index()
print(life_expectancy_le)
plt.bar(range(len(life_expectancy_le["Country"])), life_expectancy_le["Life expectancy at birth (years)"])
plt.title("Average Life Expectancy by Country")
#create your ax object here
ax = plt.subplot()
ax.set_xticks(range(len(life_expectancy_le["Country"])))
ax.set_xticklabels(life_expectancy_le["Country"], rotation=30)
plt.show()
plt.clf()

print("Although gdps are variable, the life expectancies are nearly the same except for Zimbabwe")
```

Main conclusions
      
1) Life expectancy is skewed to the left, with the peak at the upper end

2) GDP is skewed right, with majority on the low income.
      Meaning that high life expectancies are more popular despite the majority of GDP being low income.
      
The GDP only is correlated to life expectancy to a certain extent. You need only a moderate GDP improvement
to increase life expectancy to a comparible level of a first world country This shows that it does not take
a country to be in the top of the economy to have a good life expectancy. Despite variability in GDP, 
life expectancy stabilises after a certain income.
